```python
def plot_corr(df, size = 11):
    
    corr = df.corr()      #data frame correlation function
    fig, ax = plt.subplots(figsize=(size, size))
    ax.matshow(corr)      #colour code the rectangles by correlation value
    plt.xticks(range(len(corr.columns)), corr.columns)  #draw xX tick marks
    plt.yticks(range(len(corr.columns)), corr.columns)  #draw y tick marks
```

## Spliting the data

## Training - 70% , Testing - 30%

```python
from sklearn.model_selection import train_test_split

feature_col_names = ['age','anaemia','creatinine_phosphokinase', 'diabetes', 'ejection_fraction', 'high_blood_pressure', 'serum_creatinine','serum_sodium','sex','smoking','time']
prediction_class_names = ['EVENT']

X = df[feature_col_names].values
y = df[prediction_class_names].values
split_test_size = 0.30

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = split_test_size, random_state = 42)
```

#### Check to ensure we 70% training and 30% test data set

```python
print("{0:0.2f}% in training set". format((len(X_train)/len(df.index))*100))
print("{0:0.2f}% in test set". format((len(X_test)/len(df.index))*100))
```

```python
##from sklearn.preprocessing import Imputer
from sklearn.impute import SimpleImputer
# Impute with mean all 0 reading

fill_0 = SimpleImputer(missing_values =0, strategy = "mean")

X_train = fill_0.fit_transform(X_train)
X_test = fill_0.fit_transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB

nb_model = GaussianNB()

nb_model.fit(X_train, y_train.ravel())
```

## Performance on Training Data 

```python
#Predict values using the training data
nb_predict_train = nb_model.predict(X_train)

# import the performance metrics library
from sklearn import metrics

# Accuracy
print("Accuracy: {0:.4f}".format(metrics.accuracy_score(y_train, nb_predict_train)))
print()
```

## Performance on Testing Data

```python
# Predict values using the testing data
nb_predict_test = nb_model.predict(X_test)

from sklearn import metrics

# training accuracy
print("Accuracy: {0:.4f}".format(metrics.accuracy_score(y_test, nb_predict_test)))

```
